```python
import cv2
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    return image
img = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/bg.jpg")



x = 680
y = 300

center_coordinates = (x,y)

axesLength = (200,100)

angle = 0
  
startAngle = 0

endAngle = 90


ell = cv2.ellipse(img, center_coordinates, axesLength,
           angle, startAngle, endAngle, 255, 10)

plt.imshow(value(ell))

%reset -f
```

# Draw a circle with ellipse

```python
import cv2
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    return image
img = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/bg.jpg")



x = 680
y = 300

center_coordinates = (x,y)

axesLength = (100,100)

angle = 0
  
startAngle = 0

endAngle = 360


ell = cv2.ellipse(img, center_coordinates, axesLength,
           angle, startAngle, endAngle, (0,0,255), 10)

plt.imshow(value(ell))

%reset -f
```
