```python
import cv2
import matplotlib.pyplot as plt
```

```python
def convert(img):
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return image
```

```python
img1 = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/an1.jpg")
img2 = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/bg.jpg")

img1 = convert(img1)
img2 = convert(img2)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img2)
```

# resize both image

```python
img1 = cv2.resize(img1,(800,800))
img2 = cv2.resize(img2,(800,800))
```

# cv2.add(img1,img2)

```python
new = cv2.add(img1,img2)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(new)
```
