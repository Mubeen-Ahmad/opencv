# cv2.imwrite() 
## cv2.imwrite() method is used to save an image to any storage device. 

# Syntax
## cv2.imwrite(filename, image)
* ## filename: A string representing the file name. 
* ## The filename must include image format like .jpg, .png, etc.
## Return Value: It returns true if image is saved successfully.

```python
import cv2
```

```python
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15,10))
    return image
```

```python
image = cv2.imread("./px1.jpg")
```

```python
plt.imshow(value(image))
```

# Change the image

```python
value(image)
```

## Reverse Image value

```python
new = value(image)[::,::-1]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(new)
```

# write image

```python
cv2.imwrite("new.jpg", image)
```

```python
save_im = cv2.imread("new.jpg")
```

## Show new image

```python
plt.imshow(value(save_im))
```
