# 1. Read an image using imread() function.


## Syntax: 
* ## cv2.imread(path, flag)

## Parameters:
* ### path: A string representing the path of the image to be read.
* ### flag: It specifies the way in which image should be read. Itâ€™s default value is cv2.IMREAD_COLOR

### Return Value: This method returns an image that is loaded from the specified file.

```python
import cv2
image = cv2.imread("./px1.jpg",cv2.IMREAD_COLOR)
```

# Create a GUI window and display image using imshow() function.
# Syntax:
* ### cv2.imshow("Window name",your image)
$$$$
* ### cv2.waitkey(0)
    * ### To hold the window on screen, we use cv2.waitKey method
    * ### It should be positive integer. If 0 pass an parameter, then it will hold the screen until user close it.
$$$$
* ### cv2.destroyAllWindows()
    * ### It is for removing/deleting created GUI window from screen and memory


```python
cv2.imshow("window",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Note 
## In jupyter-Notebook i am used matplotlib.pyplot

```python
def show(image):
    from matplotlib import pyplot as plt
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15,10)) 
    return plt.imshow(image)
```

```python
show(image)
```

# Note: 
   * ## By default, OpenCV stores colored images in BGR(Blue Green and Red) format.

## All three types of flags are described below:

## cv2.IMREAD_COLOR
   * ### it specifies to load a color image. 
   * ### Any transparency of image will be neglected. It is the default flag. 
   * ### Alternatively, we can pass integer value 1 for this flag.

## cv2.IMREAD_GRAYSCALE
   * ### it specifies to load an image in grayscale mode. 
   * ### Alternatively, we can pass integer value 0 for this flag.

## cv2.IMREAD_UNCHANGED
   * ### it specifies to load an image as such including alpha channel. 
   * ### Alternatively, we can pass integer value -1 for this flag.

```python
color = cv2.imread("./px1.jpg",cv2.IMREAD_COLOR)
show(color)
```

```python
gray = cv2.imread("./px1.jpg",cv2.IMREAD_GRAYSCALE)

show(gray)
```

```python
unchange = cv2.imread("./px1.jpg",cv2.IMREAD_UNCHANGED)
show(unchange)
```
