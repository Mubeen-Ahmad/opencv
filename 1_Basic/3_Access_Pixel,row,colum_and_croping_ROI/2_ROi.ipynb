```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    return image
```

```python
img1 = cv2.imread("./pic/an1.jpg")
plt.imshow(value(img1))
```

# Extract FACE

```python
face = img1[180:300,200:330]
plt.imshow(value(face))
```

# Now Add Face

## Note make sure Difference are same in x (width) or y (Height)

```python
# e.g 
# difference in y is 300-180 = 120
# difference in x is 330-200 = 130


y = 150
x = 45

img1[180+y:300+y,200-x:330-x] = face

plt.imshow(value(img1))

# check difference are same
print( (300+y) - (180+y) , (330-x) - (200-x))
```
