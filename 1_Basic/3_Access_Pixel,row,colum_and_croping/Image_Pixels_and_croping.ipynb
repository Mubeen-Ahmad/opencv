```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    return image
```

```python
img1 = cv2.imread("./pic/an1.jpg",1)
plt.imshow(value(img1))
```
