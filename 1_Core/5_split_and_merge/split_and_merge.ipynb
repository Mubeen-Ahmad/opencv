```python
import cv2
import matplotlib.pyplot as plt

def convert(img):
    plt.figure(figsize=(10,10))
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.imread("../../2_Image_Processing/2_Color_Spaces/col.png")
img1 = convert(img1)
```


# Split with indexing

    img[:,:(chanel)]
     its depend on array Dim..


```python
b = img1[:,:,(0)]
g = img1[:,:,(1)]
r = img1[:,:,(2)]
```

```python
plt.imshow(convert(b))
```

```python
plt.imshow(convert(g))
```

```python
plt.imshow(convert(r))
```

# split with cv2.split()

# Note it is better to split array with index because indexing consume less time

```python
sb,sg,sr = cv2.split(img1)
```

```python
plt.imshow(convert(sb))
```

# merge with cv2.merge([arrays])

```python
new = cv2.merge([b,g,r])
```

```python
plt.imshow(new)
```

# merge with numpy np.hstack((arrays))
# merge mutiple images in one images

```python
import numpy as np
nm = np.hstack((convert(b),convert(g),convert(r)))
plt.imshow(nm)
```

# Add colors in hstack

```python
b,g,r = cv2.split(img1)

size = b.shape[0:2]
print(size)
z = np.zeros(size,dtype="uint8")
```

```python
b = cv2.merge([b,z,z])
plt.imshow(convert(b))
```

```python
g = cv2.merge([z,g,z])
plt.imshow(convert(g))
```

```python
_,_,r = cv2.split(img1)
r = cv2.merge([z,z,r])
plt.imshow(convert(r))
```

# merge

```python
b,g,r = cv2.split(img1)
cb = cv2.merge([b,z,z])
cg = cv2.merge([z,g,z])
cr = cv2.merge([z,z,r])

new = np.hstack([cb,cg,cr])
plt.imshow(convert(new))
```
