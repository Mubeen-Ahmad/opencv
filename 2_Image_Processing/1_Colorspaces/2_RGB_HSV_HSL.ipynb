# RGB

### Rgb is the simple color palette in 3 D
## [R ,G , B]


# HSV

## H = Hue Value
* ## Hue are the Complete Color palette (e.g red,blue,green)

## S = Saturation Value
* ## Saturation is the specfic color value (e.g reg ---> dark red, medium red, light red)

## V = intensity of color value
* ## Intesity is the Brightness of color (e,g dark , light)

# HSL 

## HSL are the same as HSV but there is the little difference in V and L

## in HSV v is the Intesity but In HSL L means the lighter color 

* ## L = max light(white) / min light(black)
* ## Max value of L is 50

# RGB Image

```python
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    return image
```

```python
import cv2
import numpy as np
img = cv2.imread("./col.png")

plt.imshow(value(img))
```

# Convert RGB to HSV

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV_FULL)

plt.imshow(value(hsv))
```
