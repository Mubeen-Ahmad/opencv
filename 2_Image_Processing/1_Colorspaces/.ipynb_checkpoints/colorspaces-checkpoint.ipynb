# Color spaces 
* ## Color Spaces are a way to represent the color channels present in the image that gives the image that particular hue.

* ## There are several different color spaces and each has its own significance. 
* ## Some of the popular color spaces are :
    * ## RGB (Red, Green, Blue), 
    * ## CMYK (Cyan, Magenta, Yellow, Black), 
    * ## HSV (Hue, Saturation, Value), etc

```python
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(5,5))
    return image
```

```python
import cv2

img = cv2.imread("./col.png")
plt.imshow(value(img))
```

```python
orignal = value(img)
```

# split image

## Syntax
* ## cv2.split(image)

```python
B,G,R = cv2.split(orignal)
```

# Blue

```python
plt.imshow(value(B))
```

# Green

```python
plt.imshow(value(G))
```

# Red

```python
plt.imshow(value(R))
```
