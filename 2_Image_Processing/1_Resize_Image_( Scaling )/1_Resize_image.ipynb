# Image Resize

# Syntax :

* # cv2.resize(src, dsize[, dst[, fx[, fy[, interpolation]]]])

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(5,5))
    return image
```

```python
img = "../../1_Basic/3_Access_Pixel,row,colum_and_croping/pic/an1.jpg"
img = cv2.imread(img)
```

# check height and width with .shape

# return height , width , channal


```python
img.shape
```

# fx and fy are the factors that are multiply by height and width

```python
height = 700
width = 700

fx = 5.5
fy = 5.5

new = cv2.resize(img,(height,width),fx,fy)
plt.imshow(value(new))
```

# Choice of Interpolation Method for Resizing – 

    cv2.INTER_AREA: This is used when we need to shrink an image.
    cv2.INTER_CUBIC: This is slow but more efficient.
    cv2.INTER_LINEAR: This is primarily used when zooming is required.
    cv2.INTER_NEAREST: This is used nearest values
    
<a href ="https://iq.opengenus.org/different-interpolation-methods-in-opencv/">click here for more Interpolation<a/>

```python
area = cv2.resize(img,(400,500),interpolation=cv2.INTER_AREA)
area =  value(area)
plt.imshow(area)
```
