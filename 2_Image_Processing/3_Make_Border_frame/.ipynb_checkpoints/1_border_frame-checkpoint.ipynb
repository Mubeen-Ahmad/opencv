# cv2.copyMakeBorder() 
* ## method is used to create a border around the image like a photo frame. 
 
## cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value)


   * ## src : It is the source image. 
   * ## top: It is the border width in number of pixels in top direction. 
   * ## bottom: It is the border width in number of pixels in bottom direction. 
   * ## left: It is the border width in number of pixels in left direction. 
   * ## right: It is the border width in number of pixels in right direction. 
   * ## borderType: It depicts what kind of border to be added. It is defined by flags like cv2.BORDER_CONSTANT, cv2.BORDER_REFLECT, etc dest: It is the destination image
   * ## value: It is an optional parameter which depicts color of border if border type is cv2.BORDER_CONSTANT.

   * ## Return Value: It returns an image. 

    cv2.BORDER_CONSTANT: 
                         It adds a constant colored border. 
                         The value should be given as a keyword argument
    
    cv2.BORDER_REFLECT: The border will be mirror reflection of the border elements. 
                        Suppose, if image contains letters “abcdefg” 
                        then output will be  “gfedcba|abcdefg|gfedcba“. 
    
    cv2.BORDER_REFLECT_101 
            or 
    cv2.BORDER_DEFAULT: 
                        It does the same works as cv2.BORDER_REFLECT but with slight change. 
                        Suppose, if image contains letters “abcdefgh” then output will 
                        be “gfedcb|abcdefgh|gfedcba“. 
    
    cv2.BORDER_REPLICATE: It replicates the last element. 
                          Suppose, if image contains letters “abcdefgh” then output will be 
                          “aaaaa|abcdefgh|hhhhh“. 
 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    return image

img = "../2_Image_Blur/ron.jpg"
img = cv2.imread(img)
```

```python
plt.imshow(value(img))
```

# BORDER_CONSTANT

```python
top = 80
botom = 80
left = 80
right = 80
bordertype = cv2.BORDER_CONSTANT

new = cv2.copyMakeBorder(img,top,botom,left,right,bordertype)
plt.imshow(value(new))
```

# BORDER_REFLECT

```python
top = 0
botom = 0
left = 230
right = 230
bordertype = cv2.BORDER_DEFAULT

new = cv2.copyMakeBorder(img,top,botom,left,right,bordertype)
plt.imshow(value(new))
```

# BORDER_REFLECT_101 

```python
top = 0
botom = 0
left = 0
right = 150
bordertype = cv2.BORDER_REPLICATE

new = cv2.copyMakeBorder(img,top,botom,left,right,bordertype)
plt.imshow(value(new))
```
