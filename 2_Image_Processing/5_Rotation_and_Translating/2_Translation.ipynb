# Translation
### Shift an image in coordinate space by adding a specified value to the x and y coordinate

# There are two steps of Translation
* ## Translation matix

  * ## here tx along x axis and ty along y axis
<img src="tr.png">
   
* ## Applying matrix to the image

    * ## cv2.warpAffine() are used to apply the matrix in array

```python
import cv2,numpy as np
import matplotlib.pyplot as plt

def convert(img):
    plt.figure(figsize=(10,10))
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img = cv2.imread("../../2_Image_Processing/2_Image_Blur/ron.jpg")
plt.imshow(convert(img))
```

# Translation matrix

```python
import numpy as np

tx = 100
ty = 100

metrix = np.float32([
    [1,0,tx],
    [0,1,ty]
])

metrix
```

# Appling matrix in image

    warpAffine(src, M, dsize[, dst[, flags[, borderMode[, borderValue]]]]) -> dst
    
    src - > img
    M -> array
    dsize -> h,w

```python
img.shape
```

```python
h = img.shape[0]
w = img.shape[1]
new = cv2.warpAffine(img,metrix,(w,h))
plt.imshow(convert(new))
```

# Increase size

```python
h = img.shape[0]+100
w = img.shape[1]+100
new = cv2.warpAffine(img,metrix,(w,h))
plt.imshow(convert(new))
```
