# Line
## Syntax :
* ## cv2.line(image, start_point, end_point, color, thickness)
## Understand Line 

# (start points) , (end points)
# (x1,y1) , (x2,y2)
# x1 --> start Horizontal point
# y1 --> start Vertical point

# x2 --> start Horizontal point
# y2 --> start Vertical point


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    return image
```

```python
img = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/bg.jpg")
plt.imshow(value(img))
```

# Draw Vertical line

# x1 = 0
# y1 = 0

# x2 = 0
# y2 = 600

```python
x1 = 0
y1 = 0

x2 = 0
y2 = 600 

line = cv2.line(value(img),(x1,y1),(x2,y2),255,15)
plt.imshow(line)
```
