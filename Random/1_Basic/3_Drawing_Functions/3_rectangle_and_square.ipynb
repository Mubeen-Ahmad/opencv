# Rectangle
# Syntax: 
# cv2.rectangle(image, start_point, end_point, color, thickness)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def value(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    return image
```

```python
img = cv2.imread("../3_Access_Pixel,row,colum_and_croping/pic/bg.jpg")
plt.imshow(value(img))
```

# x1 is vertical start point
# y1 is horizontal start point

# x2 is vertical end point
# y2 is horizontal end point

# Draw a Square 

```python
x1 = 600
y1 = 250


x2 = 750
y2 = 400 

line = cv2.rectangle(value(img),(x1,y1),(x2,y2),(255,0,0),10)
plt.imshow(line)
```

# Draw a rectangle

```python
x1 = 400
y1 = 150


x2 = 1000
y2 = 400 

line = cv2.rectangle(value(img),(x1,y1),(x2,y2),(255,0,0),10)
plt.imshow(line)
```
