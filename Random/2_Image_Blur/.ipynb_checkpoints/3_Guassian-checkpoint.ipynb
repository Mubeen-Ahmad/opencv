```python
import cv2,numpy as np
import matplotlib.pyplot as plt

def convert(img):
    plt.figure(figsize=(10,10))
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread("./ron.jpg")
```

# Guassian blur

    Gaussian blur is the result of blurring an image by a Gaussian function. 
    It is a widely used effect in graphics software, typically to reduce image noise and reduce detail. 
    It is also used as a preprocessing stage before applying our machine learning or deep learning models.

### GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType]]]) -> dst
# note kernal value must be odd
        if sigmaY is zero, it is set to be equal to sigmaX, 
        if both sigmas are zeros,they are computed from ksize.width and ksize.height,


```python
k = (3,3)
new = cv2.GaussianBlur(img,k,sigmaX=34,sigmaY=36)

new = np.hstack((img,new))

plt.imshow(convert(new))
```

# stretch image Horizontal

```python
k = (1,99)
new = cv2.GaussianBlur(img,k,sigmaX=0,sigmaY=0)

new = np.hstack((img,new))

plt.imshow(convert(new))
```

# stretch image vertically

```python
k = (99,1)
new = cv2.GaussianBlur(img,k,sigmaX=0,sigmaY=0)

new = np.hstack((img,new))

plt.imshow(convert(new))
```
