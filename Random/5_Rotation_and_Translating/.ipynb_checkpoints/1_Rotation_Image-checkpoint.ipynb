# Rotating an image 

## Images can be rotated to any degree clockwise or otherwise. We just need to define rotation matrix listing rotation point, degree of rotation and the scaling factor.

```python
import cv2,numpy as np
import matplotlib.pyplot as plt


def convert(img):
    plt.figure(figsize=(10,10))
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread("../2_Image_Blur/ron.jpg")
plt.imshow(convert(img))
```

# step 1 Translation to extract rotation info with getRotationMatrix2D

    getRotationMatrix2D(center, angle, scale) -> retval
    
    center = (w,h)
    angle = set angle 90,120,etc
    scale = zoom image

# Angle 2

```python
h,w = img.shape[:2]
print(h,w)
w = w
h = h

center = (w,h)
angle = 2
scale = 1

m = cv2.getRotationMatrix2D(center,angle,scale)
m
```

# Apply array into image

    warpAffine(src, M, dsize[, dst[, flags[, borderMode[, borderValue]]]]) -> dst
    
    src = image
    m = array / metrix
    dsize =  w , h

```python
# note h , w must be integer
new = cv2.warpAffine(img,m,(int(w),int(h)))
plt.imshow(convert(new))b
```

# Angle 10

```python
h,w = img.shape[:2]
print(h,w)
w = w
h = h

center = (w,h)
angle = 10
scale = 1

m = cv2.getRotationMatrix2D(center,angle,scale)
new = cv2.warpAffine(img,m,(int(w),int(h)))
plt.imshow(convert(new))
```

# Increase Background

```python


h,w = img.shape[:2]
print(h,w)
w = w
h = h

center = (w,h)
angle = 10
scale = 1

m = cv2.getRotationMatrix2D(center,angle,scale)
new = cv2.warpAffine(img,m,(int(w)*2,int(h)*2))
plt.imshow(convert(new))
```

# Divide a Background (for Fit size)

```python
h,w = img.shape[:2]
w = w / 2
h = h / 2
print(w,h)
angle = 20
scale = 1
m = cv2.getRotationMatrix2D((w,h),angle,scale)
new = cv2.warpAffine(img,m,(int(w)+int(w),int(h)+int(h)))
plt.imshow(convert(new))
```
